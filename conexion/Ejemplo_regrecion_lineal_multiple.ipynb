```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataset

value_y_list = [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4] # valores de Y
value_x1_list = [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10] # valores de X1
value_x2_list = [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68] # valores de X2
value_x3_list = [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8] # valores de X3
value_x4_list = [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12] # valores de X4

data_dict = {"Y":value_y_list,
             "X1":value_x1_list,
             "X2":value_x2_list,
             "X3":value_x3_list,
             "X4":value_x4_list}

print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

#print(df) Tabla básica
df
```

```python
import numpy as np

# Convertir el DataFrame a una matriz NumPy
matrix = df.to_numpy()

# Crear una columna de unos con el mismo número de filas que la matriz
ones_column = np.ones((matrix.shape[0], 1))

# Reemplazar la primera columna de la matriz con la columna de unos
matrix_with_ones = np.concatenate((ones_column, matrix[:, 1:]), axis=1)

# Mostrar la matriz resultante
print("Matriz original (primeras 5 filas):\n", matrix[:5])
print("\nMatriz con la primera columna reemplazada por unos (primeras 5 filas):\n", matrix_with_ones[:5])
```

```python
# Calcular la matriz transpuesta
transposed_matrix = matrix_with_ones.T

# Mostrar la matriz transpuesta
print("Matriz transpuesta (primeras 5 filas):\n", transposed_matrix[:5])
```

```python
# Extraer la columna 'Y' del DataFrame original
Y_column_df = df['Y']

# Convertir la columna 'Y' a un vector NumPy
Y_vector = Y_column_df.to_numpy()

# Mostrar el vector resultante y su forma
print("Columna 'Y' como DataFrame:\n", Y_column_df)
print("\nVector 'Y' (primeras 5 filas):\n", Y_vector[:5])
print("\nForma del vector 'Y':", Y_vector.shape)
```

```python
# Multiplicar la matriz transpuesta por la matriz original con unos
# Usamos np.dot para la multiplicación de matrices
multiplied_matrix = np.dot(transposed_matrix, matrix_with_ones)

# Mostrar la matriz resultante y su forma
print("Matriz multiplicada (primeras 5 filas y columnas):\n", multiplied_matrix[:5, :5])
print("\nForma de la matriz multiplicada:", multiplied_matrix.shape)
```

```python
# Calcular la inversa de la matriz multiplicada
# Asegúrate de que la matriz sea cuadrada e invertible para evitar errores
inverse_multiplied_matrix = np.linalg.inv(multiplied_matrix)

# Mostrar la matriz inversa resultante y su forma
print("Matriz inversa (primeras 5 filas y columnas):\n", inverse_multiplied_matrix[:5, :5])
print("\nForma de la matriz inversa:", inverse_multiplied_matrix.shape)
```

```python
import numpy as np

# Multiplicar la matriz inversa por la matriz transpuesta
step1_multiplication = np.dot(inverse_multiplied_matrix, transposed_matrix)

# Multiplicar el resultado del paso 1 por el vector Y
beta_coefficients = np.dot(step1_multiplication, Y_vector)

# Mostrar los coeficientes beta resultantes y su forma
print("Coeficientes beta:", beta_coefficients)
print("Forma de los coeficientes beta:", beta_coefficients.shape)
```
