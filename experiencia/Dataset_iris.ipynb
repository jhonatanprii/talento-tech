Load the Iris dataset and generate its numerical, categorical, and complete descriptive statistics. Summarize key observations from each statistics table.

## Cargar Dataset de Iris

### Subtask:
Cargar el conjunto de datos de Iris desde una biblioteca de Python (por ejemplo, scikit-learn o seaborn) en un DataFrame de pandas.


**Reasoning**:
The subtask requires loading the Iris dataset from `sklearn.datasets`, converting it into a pandas DataFrame, adding the species column using the target names, and then displaying the first few rows to verify the successful loading. All these steps can be performed in a single code block.



```python
import pandas as pd
from sklearn.datasets import load_iris

# 1. Cargar el dataset de Iris
iris = load_iris()

# 2. Crear un DataFrame de pandas
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 3. Añadir la columna de especies
df_iris['species'] = iris.target_names[iris.target]

# 4. Mostrar las primeras filas del DataFrame
print("Dataset de Iris cargado exitosamente:")
df_iris.head()
```

## Estadísticas Numéricas Descriptivas

### Subtask:
Calcular y mostrar las estadísticas descriptivas (conteo, media, desviación estándar, mínimo, cuartiles, máximo) para todas las columnas numéricas del dataset de Iris. Presentar los resultados en una tabla separada.


**Reasoning**:
The subtask requires calculating and displaying descriptive statistics for numerical columns. I will select the numerical columns from `df_iris` using `select_dtypes(include='number')`, then apply the `.describe()` method, and finally print the result.



```python
print("Estadísticas Descriptivas Numéricas del Dataset de Iris:")
df_iris.select_dtypes(include='number').describe()
```

### Observaciones Clave de las Estadísticas Numéricas:

*   **Longitud del sépalo (sepal length):**
    *   El promedio es de 5.84 cm, con una desviación estándar de 0.83 cm, lo que indica una variabilidad moderada.
    *   Los valores van desde 4.3 cm hasta 7.9 cm.
*   **Ancho del sépalo (sepal width):**
    *   El promedio es de 3.06 cm, con una desviación estándar menor (0.44 cm), sugiriendo menos dispersión en este atributo.
    *   Los valores oscilan entre 2.0 cm y 4.4 cm.
*   **Longitud del pétalo (petal length):**
    *   Tiene el rango más amplio de valores (1.0 cm a 6.9 cm) y la mayor desviación estándar (1.77 cm), lo que implica una gran variabilidad, probablemente debido a las diferencias entre especies.
    *   La media es de 3.76 cm.
*   **Ancho del pétalo (petal width):**
    *   Similar a la longitud del pétalo, presenta una alta variabilidad con una desviación estándar de 0.76 cm y un rango de 0.1 cm a 2.5 cm.
    *   La media es de 1.20 cm.

En general, las longitudes y anchos de los pétalos muestran una mayor variabilidad en comparación con los sépalos, lo que podría ser un indicador importante para distinguir entre las diferentes especies de Iris.

## Estadísticas Categóricas Descriptivas

### Subtask:
