## Explorar la correlación entre las variables numéricas

### Subtask:
Visualizar y cuantificar la relación entre las diferentes características del Iris (longitud/ancho del sépalo y pétalo) usando, por ejemplo, un mapa de calor (heatmap).


**Reasoning**:
To visualize and quantify the relationship between the numerical features, I will first select the relevant columns, calculate their correlation matrix, and then display it using a heatmap with annotations and a title.



```python
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación de las Características del Iris')
plt.show()
```

## Análisis de la distribución de cada variable numérica por especie

### Subtask:
Utilizar gráficos como box plots o violin plots para entender mejor cómo se distribuyen individualmente cada una de las características (sepal_length, sepal_width, petal_length, petal_width) dentro de cada especie de Iris.


**Reasoning**:
To visualize the distribution of each numerical variable across different Iris species, I will iterate through the numerical columns and create a box plot for each, showing its distribution by species.



```python
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='species', y=col)
    plt.title(f'Distribución de {col.replace("_", " ").title()} por Especie')
    plt.xlabel('Especie')
    plt.ylabel(col.replace("_", " ").title())
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

## Preparación de los datos para un modelo de aprendizaje automático

### Subtask:
Dividir el conjunto de datos en características (X) y el objetivo (y), y luego separarlos en conjuntos de entrenamiento y prueba. Esto es un paso fundamental antes de aplicar cualquier algoritmo de Machine Learning.


**Reasoning**:
To prepare the data for machine learning, I will first import the necessary function, then define the feature matrix (X) and target vector (y) based on the numerical columns and 'species' column, respectively, and finally split them into training and testing sets using a 70-30 ratio with a fixed random state for reproducibility.



```python
from sklearn.model_selection import train_test_split

X = df[numerical_cols]
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```
