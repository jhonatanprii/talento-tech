```python
#leer un csv y pasarlo a un dataframe
#https://www.kaggle.com/datasets/arshid/iris-flower-dataset/download?datasetVersionNumber=1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
df=pd.read_csv('IRIS.csv')
df.describe()
df
```

```python
#funciones
media_sepal_length=df['sepal_length'].mean()
desv_sepal_length=df['sepal_length'].std()
print(f'media_sepal_length: {media_sepal_length}:.3f')
print(f'desv_sepal_length: {desv_sepal_length}:.3f')
```

```python
plt.hist(df['sepal_length'],bins=20, color='purple', edgecolor='black')
plt.xlabel('Valores/Values')
plt.ylabel('Frecuencia/Frequency')
plt.title('Histograma sepal_length')
plt.grid(True)
plt.show()
```
